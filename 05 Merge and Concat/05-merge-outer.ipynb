```python
import pandas as pd
week1 = pd.read_csv("../data/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("../data/Restaurant - Week 2 Sales.csv")
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
week1.merge(
    week2,
    how="outer",
    on="Customer ID",
)
```

```python
outer = week1.merge(
    week2,
    how="outer",
    on="Customer ID",
    suffixes=("-week1", "-week2"),
    indicator=True,
    sort=True,
)
outer
```

```python
week1_only = outer[outer["_merge"].isin(["left_only"])]
week1_only
```

```python
week2_only = outer[outer["_merge"].isin(["right_only"])]
week2_only
```

```python
week1_only_week2_only = outer[outer["_merge"].isin(["left_only", "right_only"])]
week1_only_week2_only
```
