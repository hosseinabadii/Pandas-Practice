```python
import pandas as pd
df = pd.read_csv("../data/chicago.csv").dropna(how="all")
df["Department"] = df["Department"].astype("category")
df.head()
```

```python
last_names = df["Name"].str.split(",").str.get(0).str.strip().str.title()
df.insert(loc=1, column="Last Name", value=last_names)
df.head()
```

```python
first_names = df["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).str.title()
df.insert(loc=1, column="First Name", value=first_names)
df.head()
```
