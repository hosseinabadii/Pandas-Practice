```python
import pandas as pd
df = pd.read_csv("../data/foods.csv")
df.head(5)
```

```python
df.pivot_table(
    values="Spend",
    index="Gender",
    aggfunc="sum",
)
```

```python
df.pivot_table(
    values="Spend",
    index="Gender",
    columns="Item",
    aggfunc="sum",
)
```

```python
df.pivot_table(
    values="Spend",
    index=["Gender", "Item"],
    columns="City",
    aggfunc="sum",
)
```

```python
df.pivot_table(
    values="Spend",
    index="Gender",
    columns=["Item", "City"],
    aggfunc="sum",
)
```
