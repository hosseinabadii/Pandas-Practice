```python
import pandas as pd
df = pd.read_csv("../data/fortune1000.csv", index_col="Rank")
sectors = df.groupby(by="Sector")
df.head(3)
```

```python
sectors.agg({
    "Revenue": "sum",
    "Profits": "sum",
    "Employees": "mean",
}).head(3)
```

```python
sectors.agg({
    "Revenue": ["sum"],
    "Profits": ["sum"],
    "Employees": ["mean"],
}).head(3)
```

```python
sectors.agg({
    "Revenue": ["size", "sum", "mean"],
    "Profits": ["max"],
    "Employees": ["mean"],
}).head(3)
```

```python
sectors[["Revenue", "Profits", "Employees"]].agg(["size", "sum", "mean"]).head(3)
```
