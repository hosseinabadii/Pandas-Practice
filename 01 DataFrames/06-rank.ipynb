```python
import pandas as pd
```

```python
nba = pd.read_csv("../data/nba.csv").dropna(how="all")
nba["College"].fillna("No College", inplace=True)
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary Rank I"] = nba["Salary"].rank(ascending=False).astype("int")
nba["Salary Rank II"] = nba["Salary"].rank(ascending=False, method="dense").astype("int")
nba.sort_values("Salary", ascending=False).head(20)[["Salary", "Salary Rank I", "Salary Rank II"]]
```
